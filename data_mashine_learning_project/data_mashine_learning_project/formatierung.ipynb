```python
import pandas as pd


file_path = 'regionaleWirtschaftsstruktur.csv'
data = pd.read_csv(file_path, delimiter=';')


# Konvertiert Spalten mit Kommas als Dezimaltrennzeiche
def convert_to_numeric(column):
    return pd.to_numeric(column.str.replace(',', '.'), errors='coerce')

# Columns 'cx' and 'cy' auch zu Konvertieren
data['cx'] = convert_to_numeric(data['cx'])
data['cy'] = convert_to_numeric(data['cy'])

# sicht
print("Cleaned Dataset:")
print(data.head())

#  Save the cleaned dataset
data.to_csv('cleaned_regionaleWirtschaftsstruktur.csv', index=False)



#  Visualisierung
import matplotlib.pyplot as plt

# zb.
plt.figure(figsize=(10, 6))
data.sort_values('anzahl_niederlassungen', ascending=False).head(10).plot(
    x='GEN', y='anzahl_niederlassungen', kind='bar', legend=False
)
plt.title('Top 10 Regions by Number of Establishments')
plt.xlabel('Region')
plt.ylabel('Number of Establishments')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
